# AI Trust Study: Behavioral Analysis Prototype
This notebook demonstrates how to process and analyze the behavioral data generated by the ISSR Trust Study platform.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import os

# 1. Load Data
sample_file = 'sample_output/trust_study_P-A3X7B2K9.json'

with open(sample_file, 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data)
print(f"Loaded {len(df)} trials for Participant: {df['participant_id'].iloc[0]}")
df.head()
```

## 1. Trust Calibration: Reliance vs. Override Rates
We calculate the percentage of trials where the participant followed the AI's recommendation.

```python
reliance_rate = (df['decision'] == 'accept').mean() * 100
print(f"Overall Reliance Rate: {reliance_rate:.1f}%")

# Distribution of decisions
df['decision'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#5C8AFF', '#F87171'])
plt.title('Decision Distribution (Accept vs Reject)')
plt.ylabel('')
plt.show()
```

## 2. Latency Analysis
Using high-resolution `performance.now()` data to analyze decision-making speed.

```python
mean_latency = df['latency_ms'].mean()
print(f"Mean Response Latency: {mean_latency:.2f} ms")

plt.figure(figsize=(8, 4))
plt.bar(df['trial'], df['latency_ms'], color='#5C8AFF')
plt.xlabel('Trial #')
plt.ylabel('Latency (ms)')
plt.title('Response Latency per Trial')
plt.show()
```

## 3. Self-Reported Confidence vs. Latency
Exploring if participant confidence correlates with decision speed.

```python
plt.scatter(df['participant_confidence'], df['latency_ms'])
plt.xlabel('Participant Confidence (1-10)')
plt.ylabel('Latency (ms)')
plt.title('Confidence vs. Latency')
plt.show()
```
